## Custom scraper

```python
import pandas as pd
import numpy as np
!pip install python-slugify
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ressources/senat_parse/debats_complete.csv', sep=';')
```

```python
from multiprocessing import Pool
import requests
from bs4 import BeautifulSoup
from slugify import slugify

base_url = 'https://www.senat.fr/seances/'
all_urls = list()

def generate_urls(df):
  df = df[['datsea', 'deburl']]
  df.columns = ['date', 'uri']
  df['uri_full'] = df['uri'].str.replace('st','s')
  df['uri_full'] = df['uri_full'].str.replace('000','001')
  return base_url + df['uri_full']
    
def scrape(url):
    res = requests.get(url)
    print(res.status_code, res.url)
    with open('/content/drive/MyDrive/Colab Notebooks/ressources/senat_parse/' + slugify(url)+'.txt', "w") as f:
      f.write(res.text)

all_urls = generate_urls(df)

p = Pool(10)
p.map(scrape, all_urls)
p.terminate()
p.join()
```
